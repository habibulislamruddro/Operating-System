```python
processes = ["p1", "p2", "p3", "p4", "p5"]
arrival_times = [2, 0, 1, 3, 2]
burst_times = [3, 5, 2, 1, 6]
```

```python
process_list = list(zip(processes, arrival_times, burst_times))
```

```python
current_time = 0
completion_time = {}
turnaround_time = {}
waiting_time = {}
visited = [False] * len(processes)
```

```python
def get_next_process():
    shortest_time = float('inf')
    index = -1
    for i in range(len(process_list)):
        if (not visited[i] and
            process_list[i][1] <= current_time and
            process_list[i][2] < shortest_time):
            shortest_time = process_list[i][2]
            index = i
    return index
```

```python
while False in visited:
    next_index = get_next_process()
    if next_index == -1:
        current_time += 1
    else:
        process, arrival, burst = process_list[next_index]
        current_time += burst
        completion_time[process] = current_time
        turnaround_time[process] = completion_time[process] - arrival
        waiting_time[process] = turnaround_time[process] - burst
        visited[next_index] = True
```

```python
print("Process\tArrival Time\tBurst Time\tCompletion Time\tTurnaround Time\tWaiting Time")
for process, arrival, burst in process_list:
    print(f"{process}\t{arrival}\t\t{burst}\t\t{completion_time[process]}\t\t{turnaround_time[process]}\t\t{waiting_time[process]}")
```
