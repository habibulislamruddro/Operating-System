```python
import threading
import time
import numpy as np
```

```python
capacity = 10
buffer = np.zeros(capacity)
producer_location=0
consumer_location=0
mutex=1
empty=10
full =0
```

```python
def acquire_mutex():
  global mutex
  while mutex<=0:
    print("waiting for release mutex")
    time.sleep(1)
  mutex-=1
def release_mutex():
  global mutex
  print("releasing mutex")
  mutex+=1
def acquire_empty():
  global empty
  while empty<=0:
    print("waiting for release empty")
    time.sleep(1)
  empty-=1
def release_empty():
  global empty
  print("releasing empty")
  empty+=1
def acquire_full():
  global full
  while full<=0:
    print("waiting for release mutex")
    time.sleep(1)
  full-=1
def release_full():
  global full
  print("releasing mutex")
  full+=1
```

```python
def producer():
  global producer_location
  for i in range(20):
    acquire_empty()
    acquire_mutex()
    buffer[producer_location]=i
    producer_location+=1
    producer_location%=10
    print("Producer location",i)
    release_mutex()
    release_full()

def consumer():
  global consumer_location
  for i in range(20):
    acquire_full()
    acquire_mutex()

    print("consumer location",buffer[consumer_location])
    consumer_location+=1
    consumer_location%=10

    release_mutex()
    release_empty()
```

```python
producer = threading.Thread(target=producer)
consumer = threading.Thread(target=consumer)

consumer.start()
producer.start()

producer.join()
consumer.join()
```
