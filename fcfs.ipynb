```python
processes=["p1","p2","p3","p4","p5"]
arrival_times = [1,5,0,3,6]
burst_times = [3,2,6,1,9]
```

```python
process_list = list(zip(processes, arrival_times, burst_times))
```

```python
print(process_list)
```

```python
process_list.sort(key=lambda x: x[1])
```

```python
start_time = 0
completion_time = []
turnaround_time = []
waiting_time = []
```

```python
for process, arrival, burst in process_list:
    if start_time < arrival:
        start_time = arrival
    start_time += burst
    completion_time.append(start_time)
    turnaround_time.append(start_time - arrival)
    waiting_time.append(start_time - arrival - burst)
```

```python
sum=0
print("Process\tArrival Time\tBurst Time\tCompletion Time\tTurnaround Time\tWaiting Time")
for i, (process, arrival, burst) in enumerate(process_list):
    print(f"{process}\t{arrival}\t\t{burst}\t\t{completion_time[i]}\t\t{turnaround_time[i]}\t\t{waiting_time[i]}")


```
