```python
s=2
import time
import threading
```

```python
def wait():
  global s
  while s<=0:
    print("wait for release lock")
    time.sleep(1)
  print("enter critical section")
  s-=1
```

```python
def signal():
  global s
  print("exit critical section")
  s+=1
```

```python
def semaphore(thread_no):
  wait()
  time.sleep(2)
  print(thread_no+" in critical section")
  sum=0
  for i in range(1000):
    for j in range(1000):
      sum+=i
  print(thread_no+str(sum))
  signal()

```

```python
t1 = threading.Thread(target=semaphore,args=("t1",))
t2 = threading.Thread(target=semaphore,args=("t2",))
t3 = threading.Thread(target=semaphore,args=("t3",))
t4 = threading.Thread(target=semaphore,args=("t4",))
t5 = threading.Thread(target=semaphore,args=("t5",))

t1.start()
t2.start()
t3.start()
t4.start()
t5.start()

t1.join()
t2.join()
t3.join()
t4.join()
t5.join()
```
